```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/Users/bhaskaradhikari/freelance/data/external/geo_insights_df')
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the visualization
sns.set(style="whitegrid")

# Visualization 1: Distribution of events by country
plt.figure(figsize=(10, 6))
country_counts = df['Country'].value_counts()
sns.barplot(x=country_counts.index, y=country_counts.values, palette="viridis")
plt.title('Distribution of Events by Country')
plt.xlabel('Country')
plt.ylabel('Number of Events')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Visualization 2: Distribution of events by the top 10 cities
plt.figure(figsize=(10, 6))
city_counts = df['City'].value_counts().head(10)  # Top 10 cities
sns.barplot(x=city_counts.values, y=city_counts.index, palette="mako")
plt.title('Top 10 Cities by Number of Events')
plt.xlabel('Number of Events')
plt.ylabel('City')
plt.tight_layout()
plt.show()
```
